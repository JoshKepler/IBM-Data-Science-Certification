```python
import pandas as pd
import requests
import html5lib
from bs4 import BeautifulSoup
```

```python
url = "https://stockanalysis.com/stocks/tsla/revenue/"
html = requests.get(url)

if html != '<Response [200]>':
    print('invalid html response')
    print(html)
else:
    print('valid html response')
```

```python
soup = BeautifulSoup(html.text, 'html5lib')
```

```python
tsla_data = pd.DataFrame(columns=["Date","Revenue", "Change", "Growth"])
```

```python
# First we isolate the body of the table which contains all the information
# Then we loop through each row and find all the column values for each row
for row in soup.find("tbody").find_all('tr'):
    col = row.find_all("td")
    date = col[0].text
    revenue = col[1].text
    change = col[2].text
    growth = col[3].text
    
    # Finally we append the data of each row to the table
    tsla_data = tsla_data.append({"Date":date, "Revenue":revenue, "Change":change, "Growth":growth}, ignore_index=True)  
```
